### Importing, concatenating, and cleaning data from [Keren-Shaul 2017](https://www.sciencedirect.com/science/article/pii/S0092867417305780), data can be found [here](https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE98969)

```python
import numpy as np
import pandas as pd
import os
import developmental_index as dvp
```

```python
labels = pd.read_csv('E:\\DATA\\rna_seq_datasets\\keren-shaul2017\\labels\\GSE98969_experimental_design_f.txt', delimiter = '\t')
```

```python
%%time

### okay, so we want ot concatenate all of the files together

files = os.listdir('E:\\DATA\\rna_seq_datasets\\keren-shaul2017')

index = pd.read_csv('E:\\DATA\\rna_seq_datasets\\keren-shaul2017\\' + files[3], delimiter = '\t', index_col = 0).index

df_output = pd.DataFrame(index = index)

i = -1

txt = []

for file in files:
    if '.txt' in file:
        txt = np.append(txt, file)

mouseID_list = ['z'] * len(txt)
sampleID_list = ['z'] * len(txt)

for file in txt:
    
    i = i + 1

    df = pd.read_csv('E:\\DATA\\rna_seq_datasets\\keren-shaul2017\\' + file, delimiter = '\t')
    joined_df = df.T.join(labels[['Well_ID', 'Mouse_ID', 'Number_of_cells', 'Batch_desc']].set_index('Well_ID'))
    joined_df = joined_df.T

    sampleID_list[i] = str(np.unique(joined_df.loc['Batch_desc']))

    mouseID = joined_df.loc['Mouse_ID'][1]
    n_cells = joined_df[joined_df.index == 'Number_of_cells']

    joined_df.drop(['Mouse_ID', 'Batch_desc'], axis = 0, inplace = True)
    w = -1
    n_cells_location = joined_df.index[-1]

    columns_2_drop = []


    for col in joined_df:
        w = w + 1
        if joined_df.loc[n_cells_location].values[w] == 1:
            pass
        if joined_df.loc[n_cells_location].values[w] > 1:
            joined_df[col] = joined_df[col] / joined_df.loc['Number_of_cells'][w]
        if joined_df.loc[n_cells_location].values[w] == 0:
            columns_2_drop = np.append(columns_2_drop, col)

    joined_df.drop(n_cells_location, axis = 0, inplace = True)
    joined_df.drop(columns_2_drop, axis = 1, inplace = True)

    joined_df.rename(columns = {joined_df.columns[0]: mouseID})
    mean_df = pd.DataFrame(joined_df.mean(axis = 1))

    mean_df.rename(columns = {0:str(mouseID + '-' + str(i))}, inplace = True)

    mouseID_list[i] = mouseID

    df_final = mean_df.copy()
    df_output = df_output.join(df_final)
    
    print(df_output)
```
