```python
def batch_concatenate_excel_file(direc):
    
    #importing all of the necessary packages
    
    import numpy as np
    import pandas as pd
    import os
    from os import listdir
    from os.path import isfile, join\

    #creating a list of all files within the directory provided
    
    onlyfiles = [f for f in listdir(direc) 
                 if isfile(join(direc, f))]

    #grabbing only the csv files, and creating a list of those filenames
    
    csvs = []
    for i in range(len(onlyfiles)):
        if '.csv' in onlyfiles[i]:
            csvs = np.append(csvs, onlyfiles[i])
            
    #initializing the for loop with an empty dataframe with the correct column names
    
    x = pd.read_csv(direc + '\\' + csvs[0])
    df = pd.DataFrame(columns = x.columns)
    
    #for loop through all of the csv files listed, reading each one into a dataframe and appending it to bottom of the empty dataframe
    #also add a column containing the filename

    for i in range(len(csvs)):
        filename = csvs[i]
        ds = pd.read_csv(direc + '\\' + filename)
        ds['file'] = filename
        df = df.append(ds)
        
```
